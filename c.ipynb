```c
// Basic C Syntax Examples
#include <stdio.h>
#include <stdlib.h>

int main() {
    // Variables and data types
    int age = 25;
    float height = 5.8;
    char grade = 'A';
    char name[] = "John";
    
    printf("=== Variables ===\n");
    printf("Name: %s, Age: %d, Height: %.1f, Grade: %c\n", name, age, height, grade);
    
    // Operators
    int a = 10, b = 3;
    printf("\n=== Operators ===\n");
    printf("a = %d, b = %d\n", a, b);
    printf("Addition: %d + %d = %d\n", a, b, a + b);
    printf("Subtraction: %d - %d = %d\n", a, b, a - b);
    printf("Multiplication: %d * %d = %d\n", a, b, a * b);
    printf("Division: %d / %d = %d\n", a, b, a / b);
    printf("Modulo: %d %% %d = %d\n", a, b, a % b);
    
    // Conditional statements
    printf("\n=== Conditionals ===\n");
    if (age >= 18) {
        printf("You are an adult\n");
    } else {
        printf("You are a minor\n");
    }
    
    // Switch statement
    printf("\n=== Switch Statement ===\n");
    switch (grade) {
        case 'A':
            printf("Excellent!\n");
            break;
        case 'B':
            printf("Good!\n");
            break;
        case 'C':
            printf("Average\n");
            break;
        default:
            printf("Unknown grade\n");
    }
    
    // Loops
    printf("\n=== Loops ===\n");
    printf("For loop (1-5): ");
    for (int i = 1; i <= 5; i++) {
        printf("%d ", i);
    }
    printf("\n");
    
    printf("While loop countdown: ");
    int count = 5;
    while (count > 0) {
        printf("%d ", count);
        count--;
    }
    printf("\n");
    
    return 0;
}
```

## 2. Functions & Pointers üîó

```c
// Functions & Pointers Examples
#include <stdio.h>
#include <stdlib.h>

// Function declarations
int add(int a, int b);
void swap(int *a, int *b);
int factorial(int n);
void printArray(int arr[], int size);

// Function definitions
int add(int a, int b) {
    return a + b;
}

void swap(int *a, int *b) {
    int temp = *a;
    *a = *b;
    *b = temp;
}

int factorial(int n) {
    if (n <= 1) return 1;
    return n * factorial(n - 1);
}

void printArray(int arr[], int size) {
    printf("Array: ");
    for (int i = 0; i < size; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
}

int main() {
    // Function calls
    printf("=== Functions ===\n");
    int result = add(5, 3);
    printf("add(5, 3) = %d\n", result);
    
    printf("factorial(5) = %d\n", factorial(5));
    
    // Pointers
    printf("\n=== Pointers ===\n");
    int x = 10, y = 20;
    printf("Before swap: x = %d, y = %d\n", x, y);
    swap(&x, &y);
    printf("After swap: x = %d, y = %d\n", x, y);
    
    // Pointer basics
    int num = 42;
    int *ptr = &num;
    printf("\nnum = %d\n", num);
    printf("Address of num = %p\n", (void*)&num);
    printf("ptr = %p\n", (void*)ptr);
    printf("*ptr = %d\n", *ptr);
    
    // Array and pointer relationship
    int numbers[] = {1, 2, 3, 4, 5};
    int size = sizeof(numbers) / sizeof(numbers[0]);
    printArray(numbers, size);
    
    // Pointer arithmetic
    printf("\n=== Pointer Arithmetic ===\n");
    int *p = numbers;
    for (int i = 0; i < size; i++) {
        printf("numbers[%d] = %d, *(p + %d) = %d\n", i, numbers[i], i, *(p + i));
    }
    
    return 0;
}
```

## 4. Structures üèóÔ∏è

```c
// Structures Examples
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

// Define structures
struct Point {
    int x;
    int y;
};

struct Student {
    char name[50];
    int id;
    float gpa;
    struct Point location;  // Nested structure
};

// Function to display student info
void displayStudent(struct Student s) {
    printf("Name: %s\n", s.name);
    printf("ID: %d\n", s.id);
    printf("GPA: %.2f\n", s.gpa);
    printf("Location: (%d, %d)\n", s.location.x, s.location.y);
    printf("------------------------\n");
}

// Function to calculate distance between two points
float distance(struct Point p1, struct Point p2) {
    int dx = p1.x - p2.x;
    int dy = p1.y - p2.y;
    return sqrt(dx*dx + dy*dy);
}

int main() {
    printf("=== Structures ===\n");
    
    // Create and initialize structures
    struct Point origin = {0, 0};
    struct Point p1 = {3, 4};
    
    printf("Origin: (%d, %d)\n", origin.x, origin.y);
    printf("Point 1: (%d, %d)\n", p1.x, p1.y);
    printf("Distance: %.2f\n\n", distance(origin, p1));
    
    // Student examples
    struct Student student1;
    strcpy(student1.name, "Alice Johnson");
    student1.id = 12345;
    student1.gpa = 3.85;
    student1.location.x = 10;
    student1.location.y = 20;
    
    struct Student student2 = {"Bob Smith", 67890, 3.92, {15, 25}};
    
    printf("=== Student Records ===\n");
    displayStudent(student1);
    displayStudent(student2);
    
    // Array of structures
    struct Point polygon[4] = {
        {0, 0}, {5, 0}, {5, 3}, {0, 3}
    };
    
    printf("=== Polygon Vertices ===\n");
    for (int i = 0; i < 4; i++) {
        printf("Vertex %d: (%d, %d)\n", i+1, polygon[i].x, polygon[i].y);
    }
    
    // Pointers to structures
    printf("\n=== Pointer to Structure ===\n");
    struct Student *ptr = &student1;
    printf("Using pointer:\n");
    printf("Name: %s\n", ptr->name);
    printf("ID: %d\n", ptr->id);
    printf("GPA: %.2f\n", ptr->gpa);
    
    return 0;
}
```

## 5. Memory Management üß†

```c
// Memory Management Examples
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

int main() {
    printf("=== Dynamic Memory Allocation ===\n");
    
    // malloc - allocate memory
    int n = 5;
    int *arr = (int*)malloc(n * sizeof(int));
    
    if (arr == NULL) {
        printf("Memory allocation failed!\n");
        return 1;
    }
    
    // Initialize array
    for (int i = 0; i < n; i++) {
        arr[i] = (i + 1) * 10;
    }
    
    printf("Dynamic array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
    
    // realloc - resize memory
    printf("\n=== Resizing Array ===\n");
    n = 8;
    arr = (int*)realloc(arr, n * sizeof(int));
    
    if (arr == NULL) {
        printf("Memory reallocation failed!\n");
        return 1;
    }
    
    // Initialize new elements
    for (int i = 5; i < n; i++) {
        arr[i] = (i + 1) * 10;
    }
    
    printf("Resized array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
    
    // calloc - allocate and initialize to zero
    printf("\n=== calloc Example ===\n");
    int *zeros = (int*)calloc(5, sizeof(int));
    
    if (zeros != NULL) {
        printf("calloc array (initialized to 0): ");
        for (int i = 0; i < 5; i++) {
            printf("%d ", zeros[i]);
        }
        printf("\n");
        free(zeros);
    }
    
    // Dynamic string allocation
    printf("\n=== Dynamic String ===\n");
    char *str = (char*)malloc(50 * sizeof(char));
    
    if (str != NULL) {
        strcpy(str, "Hello, Dynamic Memory!");
        printf("Dynamic string: %s\n", str);
        printf("String length: %lu\n", strlen(str));
        free(str);
    }
    
    // 2D dynamic array
    printf("\n=== 2D Dynamic Array ===\n");
    int rows = 3, cols = 4;
    int **matrix = (int**)malloc(rows * sizeof(int*));
    
    for (int i = 0; i < rows; i++) {
        matrix[i] = (int*)malloc(cols * sizeof(int));
    }
    
    // Initialize matrix
    for (int i = 0; i < rows; i++) {
        for (int j = 0; j < cols; j++) {
            matrix[i][j] = i * cols + j + 1;
        }
    }
    
    // Print matrix
    printf("Dynamic 2D array:\n");
    for (int i = 0; i < rows; i++) {
        for (int j = 0; j < cols; j++) {
            printf("%3d ", matrix[i][j]);
        }
        printf("\n");
    }
    
    // Free 2D array
    for (int i = 0; i < rows; i++) {
        free(matrix[i]);
    }
    free(matrix);
    
    // Don't forget to free the original array!
    free(arr);
    
    printf("\n‚úÖ All memory freed successfully!\n");
    
    return 0;
}
```

```c
// SOLUTIONS - Complete implementations

// ===== SOLUTION 1: Number Analyzer =====
#include <stdio.h>
#include <stdlib.h>

int findMax(int arr[], int size) {
    int max = arr[0];
    for (int i = 1; i < size; i++) {
        if (arr[i] > max) {
            max = arr[i];
        }
    }
    return max;
}

int countEven(int arr[], int size) {
    int count = 0;
    for (int i = 0; i < size; i++) {
        if (arr[i] % 2 == 0) {
            count++;
        }
    }
    return count;
}

void reverseArray(int arr[], int size) {
    for (int i = 0; i < size / 2; i++) {
        int temp = arr[i];
        arr[i] = arr[size - 1 - i];
        arr[size - 1 - i] = temp;
    }
}

// ===== SOLUTION 2: String Processor =====
int countVowels(char str[]) {
    int count = 0;
    for (int i = 0; str[i] != '\0'; i++) {
        char c = str[i];
        if (c == 'a' || c == 'e' || c == 'i' || c == 'o' || c == 'u' ||
            c == 'A' || c == 'E' || c == 'I' || c == 'O' || c == 'U') {
            count++;
        }
    }
    return count;
}

void removeSpaces(char str[]) {
    int writePos = 0;
    for (int readPos = 0; str[readPos] != '\0'; readPos++) {
        if (str[readPos] != ' ') {
            str[writePos] = str[readPos];
            writePos++;
        }
    }
    str[writePos] = '\0';
}

int isPalindrome(char str[]) {
    int len = strlen(str);
    for (int i = 0; i < len / 2; i++) {
        if (str[i] != str[len - 1 - i]) {
            return 0;
        }
    }
    return 1;
}

int countWords(char str[]) {
    int count = 0;
    int inWord = 0;
    for (int i = 0; str[i] != '\0'; i++) {
        if (str[i] != ' ') {
            if (!inWord) {
                count++;
                inWord = 1;
            }
        } else {
            inWord = 0;
        }
    }
    return count;
}

// ===== SOLUTION 3: Student Record System =====
typedef struct {
    int id;
    char name[50];
    float grades[5];
    int numGrades;
} Student;

void addGrade(Student *student, float grade) {
    if (student->numGrades < 5) {
        student->grades[student->numGrades] = grade;
        student->numGrades++;
    } else {
        printf("Cannot add more grades (maximum 5 reached)\n");
    }
}

float calculateAverage(Student *student) {
    if (student->numGrades == 0) return 0.0;
    
    float sum = 0;
    for (int i = 0; i < student->numGrades; i++) {
        sum += student->grades[i];
    }
    return sum / student->numGrades;
}

int findTopStudent(Student students[], int count) {
    int topIndex = 0;
    float topAverage = calculateAverage(&students[0]);
    
    for (int i = 1; i < count; i++) {
        float avg = calculateAverage(&students[i]);
        if (avg > topAverage) {
            topAverage = avg;
            topIndex = i;
        }
    }
    return topIndex;
}

void displayStudent(Student *student) {
    printf("ID: %d\n", student->id);
    printf("Name: %s\n", student->name);
    printf("Grades: ");
    for (int i = 0; i < student->numGrades; i++) {
        printf("%.1f ", student->grades[i]);
    }
    printf("\nAverage: %.2f\n", calculateAverage(student));
}

// Test function to demonstrate all solutions
int testSolutions() {
    printf("=== TESTING ALL SOLUTIONS ===\n\n");
    
    // Test number analyzer
    int numbers[] = {12, 7, 23, 8, 15, 4, 19, 6};
    int size = 8;
    
    printf("1. Number Analyzer:\n");
    printf("Array: ");
    for (int i = 0; i < size; i++) printf("%d ", numbers[i]);
    printf("\nMax: %d, Even count: %d\n", findMax(numbers, size), countEven(numbers, size));
    
    reverseArray(numbers, size);
    printf("Reversed: ");
    for (int i = 0; i < size; i++) printf("%d ", numbers[i]);
    printf("\n\n");
    
    // Test string processor
    char text[] = "Hello World";
    printf("2. String Processor:\n");
    printf("Text: \"%s\"\n", text);
    printf("Vowels: %d, Words: %d\n", countVowels(text), countWords(text));
    printf("\"racecar\" is palindrome: %s\n\n", isPalindrome("racecar") ? "Yes" : "No");
    
    // Test student system
    Student students[2] = {
        {101, "Alice", {85.5, 92.0, 78.5}, 3},
        {102, "Bob", {90.0, 88.5, 95.0}, 3}
    };
    
    printf("3. Student System:\n");
    for (int i = 0; i < 2; i++) {
        printf("Student %d: %s, Average: %.2f\n", 
               i+1, students[i].name, calculateAverage(&students[i]));
    }
    printf("Top student: %s\n", students[findTopStudent(students, 2)].name);
    
    printf("\n‚úÖ All solutions working correctly!\n");
    return 0;
}
```

```c
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <string.h>

// Define neural network structure
typedef struct {
    float *data;
    int rows;
    int cols;
} Matrix;

typedef struct {
    Matrix conv1_weights;
    Matrix conv1_bias;
    Matrix conv2_weights;
    Matrix conv2_bias;
    Matrix fc1_weights;
    Matrix fc1_bias;
    Matrix fc2_weights;
    Matrix fc2_bias;
} MNISTModel;

// Matrix operations
Matrix create_matrix(int rows, int cols) {
    Matrix m;
    m.rows = rows;
    m.cols = cols;
    m.data = (float*)malloc(rows * cols * sizeof(float));
    return m;
}

void free_matrix(Matrix *m) {
    if (m->data) {
        free(m->data);
        m->data = NULL;
    }
}

// Activation functions
float relu(float x) {
    return x > 0 ? x : 0;
}

float softmax(float *input, int size, int index) {
    float sum = 0.0;
    for (int i = 0; i < size; i++) {
        sum += exp(input[i]);
    }
    return exp(input[index]) / sum;
}

// Convolution operation (simplified)
void conv2d(float *input, int input_h, int input_w, int input_c,
           float *kernel, int kernel_h, int kernel_w, int out_c,
           float *bias, float *output) {
    int output_h = input_h - kernel_h + 1;
    int output_w = input_w - kernel_w + 1;
    
    for (int oc = 0; oc < out_c; oc++) {
        for (int oh = 0; oh < output_h; oh++) {
            for (int ow = 0; ow < output_w; ow++) {
                float sum = 0.0;
                for (int ic = 0; ic < input_c; ic++) {
                    for (int kh = 0; kh < kernel_h; kh++) {
                        for (int kw = 0; kw < kernel_w; kw++) {
                            int input_idx = ic * input_h * input_w + 
                                          (oh + kh) * input_w + (ow + kw);
                            int kernel_idx = oc * input_c * kernel_h * kernel_w + 
                                           ic * kernel_h * kernel_w + 
                                           kh * kernel_w + kw;
                            sum += input[input_idx] * kernel[kernel_idx];
                        }
                    }
                }
                int output_idx = oc * output_h * output_w + oh * output_w + ow;
                output[output_idx] = relu(sum + bias[oc]);
            }
        }
    }
}

// Max pooling operation
void max_pool2d(float *input, int input_h, int input_w, int channels,
               int pool_size, float *output) {
    int output_h = input_h / pool_size;
    int output_w = input_w / pool_size;
    
    for (int c = 0; c < channels; c++) {
        for (int oh = 0; oh < output_h; oh++) {
            for (int ow = 0; ow < output_w; ow++) {
                float max_val = -INFINITY;
                for (int ph = 0; ph < pool_size; ph++) {
                    for (int pw = 0; pw < pool_size; pw++) {
                        int input_idx = c * input_h * input_w + 
                                      (oh * pool_size + ph) * input_w + 
                                      (ow * pool_size + pw);
                        if (input[input_idx] > max_val) {
                            max_val = input[input_idx];
                        }
                    }
                }
                int output_idx = c * output_h * output_w + oh * output_w + ow;
                output[output_idx] = max_val;
            }
        }
    }
}

// Fully connected layer
void fully_connected(float *input, int input_size, 
                     float *weights, float *bias, int output_size,
                     float *output) {
    for (int i = 0; i < output_size; i++) {
        float sum = 0.0;
        for (int j = 0; j < input_size; j++) {
            sum += input[j] * weights[i * input_size + j];
        }
        output[i] = sum + bias[i];
    }
}

// Load model weights from file (simplified binary format)
int load_model_weights(const char *filepath, MNISTModel *model) {
    FILE *file = fopen(filepath, "rb");
    if (!file) {
        printf("‚ùå Error: Cannot open model file %s\n", filepath);
        return 0;
    }
    
    printf("üîÑ Loading model weights from %s...\n", filepath);
    
    // Initialize model matrices (simplified - in real implementation, 
    // you'd read the actual dimensions from the file)
    model->conv1_weights = create_matrix(32 * 1 * 3 * 3, 1);
    model->conv1_bias = create_matrix(32, 1);
    model->conv2_weights = create_matrix(64 * 32 * 3 * 3, 1);
    model->conv2_bias = create_matrix(64, 1);
    model->fc1_weights = create_matrix(128, 9216);
    model->fc1_bias = create_matrix(128, 1);
    model->fc2_weights = create_matrix(10, 128);
    model->fc2_bias = create_matrix(10, 1);
    
    // In a real implementation, you would read the actual weights from file
    // For demo purposes, initialize with random values
    printf("‚ö†Ô∏è  Note: Using random weights for demonstration\n");
    printf("   In real implementation, weights would be loaded from file\n");
    
    // Initialize with small random values
    srand(42);
    for (int i = 0; i < model->conv1_weights.rows * model->conv1_weights.cols; i++) {
        model->conv1_weights.data[i] = ((float)rand() / RAND_MAX) * 0.1f - 0.05f;
    }
    
    fclose(file);
    printf("‚úÖ Model structure initialized\n");
    return 1;
}

// Forward pass through the network
void mnist_forward(MNISTModel *model, float *input, float *output) {
    // Input: 28x28x1 image
    // Conv1: 28x28x1 -> 26x26x32
    float *conv1_out = (float*)malloc(26 * 26 * 32 * sizeof(float));
    
    // Conv2: 26x26x32 -> 24x24x64
    float *conv2_out = (float*)malloc(24 * 24 * 64 * sizeof(float));
    
    // MaxPool: 24x24x64 -> 12x12x64
    float *pool_out = (float*)malloc(12 * 12 * 64 * sizeof(float));
    
    // Flatten: 12x12x64 -> 9216
    float *flattened = pool_out; // Same memory, different interpretation
    
    // FC1: 9216 -> 128
    float *fc1_out = (float*)malloc(128 * sizeof(float));
    
    // FC2: 128 -> 10
    float *fc2_out = (float*)malloc(10 * sizeof(float));
    
    printf("üß† Running forward pass...\n");
    
    // Simplified forward pass (actual implementation would be more complex)
    printf("   Conv1: 28x28x1 -> 26x26x32\n");
    printf("   Conv2: 26x26x32 -> 24x24x64\n");
    printf("   MaxPool: 24x24x64 -> 12x12x64\n");
    printf("   Flatten: 12x12x64 -> 9216\n");
    printf("   FC1: 9216 -> 128\n");
    printf("   FC2: 128 -> 10\n");
    
    // For demonstration, create dummy output
    for (int i = 0; i < 10; i++) {
        output[i] = ((float)rand() / RAND_MAX) * 2.0f - 1.0f;
    }
    
    // Apply softmax to get probabilities
    printf("\n� Output probabilities:\n");
    for (int i = 0; i < 10; i++) {
        float prob = softmax(output, 10, i);
        printf("   Class %d: %.4f\n", i, prob);
    }
    
    // Find predicted class
    int predicted = 0;
    for (int i = 1; i < 10; i++) {
        if (output[i] > output[predicted]) {
            predicted = i;
        }
    }
    printf("\nüéØ Predicted class: %d\n", predicted);
    
    // Cleanup
    free(conv1_out);
    free(conv2_out);
    free(pool_out);
    free(fc1_out);
    free(fc2_out);
}

// Main function
int main() {
    printf("� MNIST Model Loader in C\n");
    printf("==========================\n\n");
    
    MNISTModel model;
    
    // Load model weights
    if (!load_model_weights("/models/mnist_model.pth", &model)) {
        printf("‚ùå Failed to load model weights\n");
        return 1;
    }
    
    // Create dummy input (28x28 image)
    float *input = (float*)malloc(28 * 28 * sizeof(float));
    for (int i = 0; i < 28 * 28; i++) {
        input[i] = ((float)rand() / RAND_MAX); // Random pixel values
    }
    
    printf("\nüì• Input image: 28x28 grayscale\n");
    printf("   First few pixel values: ");
    for (int i = 0; i < 5; i++) {
        printf("%.3f ", input[i]);
    }
    printf("...\n");
    
    // Run inference
    float output[10];
    mnist_forward(&model, input, output);
    
    // Cleanup
    free(input);
    free_matrix(&model.conv1_weights);
    free_matrix(&model.conv1_bias);
    free_matrix(&model.conv2_weights);
    free_matrix(&model.conv2_bias);
    free_matrix(&model.fc1_weights);
    free_matrix(&model.fc1_bias);
    free_matrix(&model.fc2_weights);
    free_matrix(&model.fc2_bias);
    
    printf("\n‚úÖ Model inference completed successfully!\n");
    return 0;
}
```
